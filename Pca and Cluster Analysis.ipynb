# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
```

```python
x.shape
```

# First applying 10 components

```python
pc=PCA(n_components=10)
```

```python
pc.fit(x_scaled)
```

```python
pc.explained_variance_ratio_
```

```python
import numpy as np
```

# Explained Variance Plot

```python
plt.figure(figsize=(12,8))
plt.plot(np.cumsum(pc.explained_variance_ratio_))
plt.xlabel('No of Components')
plt.ylabel('Emplained Variance Ratio')
plt.title('Explained Variance Ratio vs No of Components')
plt.show()
```

```python
pc.get_params()
```

```python
pc.explained_variance_
```

# Eigen Value Plot

```python
plt.figure(figsize=(12,8))
plt.plot(pc.explained_variance_)
plt.ylim([0,14])
plt.xlim([0,10])
plt.xlabel('No of Components')
plt.ylabel('Eigen Value')
plt.title('No of Components vs Eigen Value')
plt.show()
```

# From Kaiser Rule 6 Components are selected

```python
PC=PCA(n_components=6)
```

```python
x_pc_6=PC.fit_transform(x)
```

```python
x_pc_6
```

```python
PC.fit(x)
```

```python
eigen_values=pd.Series(PC.explained_variance_,index=range(1,7))
```

# 2 .Bar plot showing Eigen Values of top 3 components

```python
fig=eigen_values.iloc[:3].plot(kind='bar',grid=True,figsize=(12,8))
fig.set_xlabel('Top 3 components')
fig.set_ylabel('Eigen Value')
fig.set_xticklabels(('PCA1','PCA2','PCA3'),rotation='horizontal')
plt.show()
```

```python
clusters_centroids=[]
scores=[]
for i in range(1,10):
    knn=KMeans(n_clusters=2,init='random')
    knn.fit(x)
    score=davies_bouldin_score(x,knn.labels_)
    scores.append(score)
    #clusters_centroids.append(knn.object)
    
```

```python
mim_3=min(scores)
```

```python
scores=[]
for i in range(1,10):
    knn=KMeans(n_clusters=4,init='random')
    knn.fit(x)
    score=davies_bouldin_score(x,knn.labels_)
    scores.append(score)
```
